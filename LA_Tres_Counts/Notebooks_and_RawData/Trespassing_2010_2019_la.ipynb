```python
# Import 3rd party libraries
import json
import numpy as np
import pandas as pd
import geopandas as gpd


# Configure Notebook
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

# Read CSV
trespass = pd.read_csv('Trespassing_Data.csv')
neighbourhoods = gpd.read_file('Neighborhoods.geojson')

# Cleaning
trespass = trespass[['Date Rptd','AREA NAME','Crm Cd', 'Crm Cd Desc','Premis Desc', 'LAT', 'LON']]
trespass = trespass[trespass['Premis Desc'] == 'SINGLE FAMILY DWELLING'].reset_index()
trespass = gpd.GeoDataFrame(
    trespass, geometry=gpd.points_from_xy(trespass['LON'], trespass['LAT']))

neighbourhoods['name'] = neighbourhoods['name'].str[:-2]

# Count Trespassing
neighbourhoods['trespass_count'] = neighbourhoods['geometry'].apply(lambda row: trespass['geometry'].within(row).sum())

# to_CSV
neighbourhoods_tres_clean = neighbourhoods
path='C:\\Users\\hvill\\Destop\\
neighbourhoods_tres_clean.to_csv
```
